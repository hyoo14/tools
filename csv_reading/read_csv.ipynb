```python
import csv
```

```python
word1 = []
word2 = []
kor_score = []
```

```python
# .tsv 읽기
f = open('w353.tsv', 'r')#, encoding='utf-8')
rdr = csv.reader(f, delimiter='\t') ####tab 으로 구분되는 tsv 파일이니깐 delimiter를 지정해 준다(텝으로)

for line in rdr:
    #elements = line.split(',')
    #print(line)
    word1.append(line[0])
    word2.append(line[1])
    kor_score.append(line[2])
    
    

#r = list(rdr)
#print("Id=%s : Name=%s" % (r[0][0], r[0][1]))
 
f.close()
```
